```python
UNKNOWN_COUNTY_FIPS = 0
STATE_FIPS_MULTIPLIER = 1000
```

```python
def get_state_fips(county_fips):
    county_fips_string = str(county_fips)
    state_fips_string = county_fips_string[:-3]
    state_fips_int = int(state_fips_string)
    return state_fips_int
```

```python
state_fips_int_series = State_FIPS_min_series.apply(get_state_fips)
```

```python
def get_unknown_fips(state_abbr):
    county_code = UNKNOWN_COUNTY_FIPS
    state_code = state_fips_int_series[state_abbr] 
    fips_code = (state_code * STATE_FIPS_MULTIPLIER) + county_code
    return fips_code
```
